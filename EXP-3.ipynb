```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Electric_Production.csv", parse_dates=["DATE"], index_col="DATE")
```

```python
data.fillna(method="ffill", inplace=True)
data.drop_duplicates(inplace=True)
```

```python
monthly_data = data.resample("M").mean()
plt.figure(figsize=(10, 5))
plt.plot(monthly_data, color="cyan")
plt.title("Monthly Average Electric Production")
plt.xlabel("Date")
plt.ylabel("Average Production")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
data.hist(bins=30, color="red", alpha=0.7)
plt.title("Distribution of Electric Production")
plt.xlabel("Production Value")
plt.ylabel("Frequency")
plt.show()

```
